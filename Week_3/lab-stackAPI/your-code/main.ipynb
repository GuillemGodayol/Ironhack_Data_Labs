#### Import the necessary libraries here:

```python
# your code here
import pandas as pd 
from stackapi import StackAPI
```

#### Question 1: Find the questions and answers of last month.

```python
# I assign different queries to each endpoint
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1570492800, todate=1573171200, order='desc', sort='creation')
answers = SITE.fetch('answers', fromdate=1570492800, todate=1573171200, order='desc',sort='creation')
```

```python
print(type(questions))
```

```python
# Using the method 'get' we get the values by key 'title' inside all cells into column items
questions_title = [x.get('title') for x in questions['items']]
```

```python
questions_title
```

```python
print(len(questions_title))
```

```python
# As the length of questions is 500, I assume that the API returns just 500 results for query.
```

```python
# I will check now the answers

# as answers don't have title, I will retrieve them by answer id

answers_id = [x.get('answer_id') for x in answers['items']]
```

```python
print(len(answers_id))
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# I look for questions, defining a date by timestamp date and tagget with python
voted_questions = SITE.fetch('questions', fromdate=1473257600, todate=1573171200, tagged='python', min=5, order='desc', sort='creation')
```

```python
voted_questions
```

```python
# I consider that I just need the title of the question and its score
# I use the method get in list comprehension and make a list of two lists.
voted_questions_title = [[x.get('title') for x in voted_questions['items']],[x.get('score') for x in voted_questions['items']]]
```

```python
# I create a dataframe with the two lists above
df_votedQuestions = pd.DataFrame(voted_questions_title)
```

```python
# as the DF is the lists as rows, I traspose it.
df_votedQuestions = pd.DataFrame.transpose(df_votedQuestions)
```

```python
# I change the name of the columns to be more clear
df_votedQuestions = df_votedQuestions.rename(columns={0 :'Question', 1:'Score'})
```

```python
# I use loc to localize the questions that have at least score 5
df_votedQuestions.loc[df_votedQuestions['Score'] >= 5]
```
