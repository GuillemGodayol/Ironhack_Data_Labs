```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlalchemy
import seaborn as sns
from scipy import stats
import math
```

```python
cholera = pd.read_csv("../data/cholera-clean.csv", index_col=0)
```

```python
cholera.head()
```

```python
sns.catplot(x="Feces_consistency", y="Gastric_acid_index", kind="box", data=cholera, order=["Normal", "Watery", "Rice Water"])
plt.show()
```

```python
sns.catplot(x="Feces_consistency", y="Gastric_acid_index", kind="swarm", data=cholera, order=["Normal", "Watery", "Rice Water"])
plt.savefig("Gastric_acid_index vs. Feces_consistency.jpg")
plt.show()
```

```python
sns.catplot(x="Feces_consistency", y="Gastric_acid_index", kind="violin", data=cholera, order=["Normal", "Watery", "Rice Water"])
plt.show()
```

```python
sns.catplot(x="Vomit_color", y="Gastric_acid_index", kind="box", data=cholera)
plt.show()
```

```python
sns.catplot(x="Vomit_color", y="Gastric_acid_index", kind="swarm", data=cholera)
plt.savefig("Gastric_acid_index vs. Vomit_color.jpg")
plt.show()
```

```python
cholera['Vomit_color'].replace({"Normal":0, "Abnormal":1, "Clear":2}, inplace = True)
cholera['Vomit_color'] = cholera['Vomit_color'].astype('int')
cholera['Feces_consistency'].replace({"Normal":0,"Watery":1, "Rice Water":2}, inplace = True)
cholera['Feces_consistency'] = cholera['Feces_consistency'].astype('int')
```

```python
A = cholera.corr()
f, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(A, annot=True, cmap='Blues')
plt.savefig("Heatmap.jpg")
plt.show()
```
