####  What is the average purchase price?

```python
# To calculate the average purchase price we need to group our rows by invoice number, and we sum the columns.

grouped = orders.groupby(['InvoiceNo']).sum()
```

```python
# The average purchase price will be the mean of the column 'amount_spent'
grouped['amount_spent'].mean()
# Average purchase price = 480,76
```

#### What are the highest and lowest purchase prices? 

```python
# Calling the describe method from the amount_spent on previous grouped dataframe, will get the max and the min.
grouped['amount_spent'].describe()
# Max purchase price = 168.469
# Min purchase price = 0
```

```python
# Checking if the max can be a real purchase or may be an error.
grouped.loc[grouped['amount_spent']== max(grouped['amount_spent'])]
```

```python
# Checking how many rows have amount spent = 0. It may be a lack of data in those columns:
grouped.loc[grouped['amount_spent']== min(grouped['amount_spent'])]
```

#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
# Localizing orders with 'Spain' in 'Country' column and then grouping by CustomerID to get customers, not orders.
spanish_cust = orders.loc[orders['Country']=='Spain'].groupby('CustomerID').sum()
```

#### How many customers do we have in Spain?

```python
# Considering the previously created df with Spanish customers in each row, the describe method gives us the answer:
spanish_cust.describe()
# There are 30 customers from Spain
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# First we group the dataframe by customer and sum all the columns.
orders.groupby('CustomerID').sum().query('Quantity > 50')
```

#### Select orders from Spain that include more than 50 items.

```python
# Using query we get the orders that are from Spain.
# We group these orders by InvoiceNum, as each invoice num is assigned to an order, summing the quanitities.
# Using query again, we select those with Quantity bigger than 50.
orders.query('Country == "Spain"').groupby('InvoiceNo').sum().query('Quantity > 50')
```

#### Select all free orders.

```python
# Looking amount_spent = 0 with query method.
orders.query('amount_spent == 0')
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
# Using str.startswith to find the rows wich description starts with 'lunch bag'
orders.loc[orders['Description'].str.startswith('lunch bag')]
```

#### Select all `lunch bag` orders made in 2011.

```python
# Using the previous statement, and adding a query for the year.
orders.loc[orders['Description'].str.startswith('lunch bag')].query('year == 2011')
```
